```python
import sqlite3    #importing the sql
```

```python
#creating database
db=sqlite3.connect("IMDB_DATABASE.db")
cur=db.cursor()
```

```python
#dropping any table having the same name
cur.execute("drop table if exists imdb_1")
db.commit()
```

```python
#giving the column names and type so that csv file can be imported
cur.execute("Create Table if not exists imdb_1(sno INT, movie_names text, duration varchar, director_1 varchar, year varchar, ratings real, metascore int)")
```

```python
#opening the csv file using a loop
with open("IMDB_1.csv",'r') as file:
    total_records=0
    for i in file:
        values=[value.strip() for value in i.split(",")]
        cur.execute("insert into imdb_1 values(?,?,?,?,?,?,?)",values )
        db.commit()    #saving the record
        total_records+=1
print("Total",total_records,"inserted")
```

```python
#Display all the details of movies created by directors Christopher and Matt Reeves.
directors=['Christopher Nolan','Matt Reeves']
result="SELECT * FROM imdb_1 WHERE director_1 IN ({})".format(','.join('?' for _ in directors))
cur.execute(result,(directors),)
movies=cur.fetchall()   #fetching all the rows matching the query
for i in movies:    #printing the result
    print(i)
```

```python
db=sqlite3.connect("IMDB_DATABASE_2.db")
cur=db.cursor()
```

```python
cur.execute("drop table if exists imdb_2")
db.commit()
```

```python
#creating an empty table to fit in the values of the csv file
cur.execute("create table if not exists imdb_2(Genre1 text, Genre2 text, Genre3 text, Movies text, Collection_in_million real, Votes int,Certificate VARCHAR, Star1 VARCHAR, Star2 VARCHAR, Star3 VARCHAR, Star4 VARCHAR)")
```

```python
#importing the csv file into sql database
with open("IMDB_2.csv",'r',encoding='utf-8') as file:
    total_records=0
    for i in file:
        values=[value.strip() for value in i.split(",")]
        cur.execute("insert into imdb_2 values(?,?,?,?,?,?,?,?,?,?,?)",values )
        db.commit()    #saving the record
        total_records+=1
print("Total",total_records,"inserted")
```

```python
#Display all details of the movie with the highest number of votes
query = """SELECT Movies, star1, star2, star3, star4, votes, Genre1, Genre2, Genre3, Collection_in_million, Certificate
    FROM imdb_2 ORDER BY votes DESC"""
cur.execute(query)

# Fetch all rows including the first row with column names
all_rows = cur.fetchall()

# Check if there are any rows
if all_rows:
    # Skip the first row with column names
    data_row = all_rows[1]

    # Display the results
    print("The details of the movie with highest votes is as follows:")
    print("Star1", data_row[3])
    print("Genre1:", data_row[7])
    print("Genre2:", data_row[8])
    print("Certification:", data_row[10])
    print("VOTES:", data_row[5])
    print("Movie_name:", data_row[0])
    print("Star3:", data_row[1])
    print("Star2:", data_row[2])
    print("Genre3:", data_row[6])
else:
    print("No rows found in the table.")
```

```python
#Display movie names with gross collections in descending order
result=cur.execute("select * from imdb_2 order by Collection_in_million DESC")
movie=result.fetchall()
for i in movie:
    print(i[3])
```

```python
#Display all details of movies with comedy and action genres.
query = """
    SELECT *
    FROM imdb_2
    WHERE (Genre1 LIKE '%Comedy%' AND Genre2 LIKE '%Action%')
       OR (Genre1 LIKE '%Action%' AND Genre2 LIKE '%Comedy%')
       OR (Genre1 LIKE '%Comedy%' AND Genre3 LIKE '%Action%')
       OR (Genre1 LIKE '%Action%' AND Genre3 LIKE '%Comedy%')
       OR (Genre2 LIKE '%Comedy%' AND Genre3 LIKE '%Action%')
       OR (Genre2 LIKE '%Action%' AND Genre3 LIKE '%Comedy%')
"""
cur.execute(query)
movies_with_genres = cur.fetchall()
if movies_with_genres:
    print("Movies with Comedy and Action Genres:")
    for movie in movies_with_genres:
        print(movie)
else:
    print("No movies found with both Comedy and Action genres.")
```

# Sub_Query

```python
#Display all details from both tables where movie names are the same
result=cur.execute("select imdb_1.*,imdb_2.* from imdb_1 INNER JOIN imdb_2 on imdb_1.movie_names=imdb_2.Movies")
details=result.fetchall()
for i in details:
    print(i)
```

```python
#Display all movie names and gross collection where the genre is action.
# Query to select movie details from imdb_1 table with genre "Action"
#Here we are using ''(null values) to match the columns of both tables
query_imdb1 = """
    SELECT movie_names, director_1 as director, ratings, '' as Collection_in_million, '' as Genre1, '' as Genre2, '' as Genre3
    FROM imdb_1
    WHERE Genre1 LIKE '%Action%'
       OR Genre2 LIKE '%Action%'
       OR Genre3 LIKE '%Action%'
"""
# Query to select movie details from imdb_2 table with genre "Action"
query_imdb2 = """
    SELECT Movies as movie_names, "" as director,'' as ratings, CAST(Collection_in_million as real), Genre1, Genre2, Genre3 FROM imdb_2
    WHERE Genre1 LIKE '%Action%'
       OR Genre2 LIKE '%Action%'
       OR Genre3 LIKE '%Action%'
"""

# Combine the results of both queries using UNION ALL
query_combined = f"{query_imdb1} UNION ALL {query_imdb2}"

result_combined = cur.execute(query_combined).fetchall()

for movie in result_combined:
    print("Movie Name:", movie[0])
    print("Gross Collection:", movie[3])
    print("Genre1:", movie[4])
    print("*************************************")
```

```python
# Display all details from both tables with the highest gross collection.
#movie with the highest gross collection from imdb_1 table
query_imdb1 = """
    SELECT movie_names, duration, director_1, year, ratings, metascore, '' as Genre1, '' as Genre2, '' as Genre3,
    '' as Collection_in_million, '' as Votes, '' as Certificate FROM imdb_1 ORDER BY ratings DESC LIMIT 1"""

#movie with the highest gross collection from imdb_2 table
#here we are using CAST() to change the originl column type to our desired type, which is from text/string to real in my case
query_imdb2 = """SELECT Movies as movie_names, '' as duration, '' as director_1, '' as year, '' as ratings, '' as metascore, Genre1, Genre2, Genre3, Collection_in_million, Votes, Certificate
    FROM imdb_2 ORDER BY CAST(Collection_in_million AS REAL) DESC LIMIT 1
"""

# Executing the queries
cur.execute(query_imdb1)
cur.execute(query_imdb2)
second_row_imdb1 = cur.fetchone()
second_row_imdb2 = cur.fetchone()

# Comparing the gross collections from both tables
highest_gross_collection = None
if second_row_imdb1 and second_row_imdb2:
    if (second_row_imdb2[9]) > (second_row_imdb1[9]):
        highest_gross_collection = second_row_imdb2
    else:
        highest_gross_collection = second_row_imdb1
elif second_row_imdb1:
    highest_gross_collection = second_row_imdb1
elif second_row_imdb2:
    highest_gross_collection = second_row_imdb2

# Check if there is any movie with the highest gross collection
if highest_gross_collection:
    # Print the details of the movie with the highest gross collection
    print("Movie Name:", highest_gross_collection[0])
    print("Genre1:", highest_gross_collection[6])
    print("Genre2:", highest_gross_collection[7])
    print("Genre3:", highest_gross_collection[8])
    print("Gross Collection(in million):", highest_gross_collection[9])
    print("Votes:", highest_gross_collection[10])
    print("Certificate:", highest_gross_collection[11])
else:
    print("No movie with the highest gross collection found.")
```

```python
#Display the details of movies having highest rating
#movie with the highest rating from imdb_1 table
query_imdb1 = """SELECT movie_names, duration, director_1, year, ratings, metascore, '' as Genre1, '' as Genre2, '' as Genre3,
    '' as Collection_in_million, '' as Votes, '' as Certificate FROM imdb_1 WHERE ratings != '' ORDER BY CAST(ratings AS REAL) DESC
    LIMIT 1"""

# Get the movie with the highest rating from imdb_2 table
query_imdb2 = """SELECT Movies as movie_names, '' as duration, '' as director_1, '' as year,'' as ratings, '' as metascore,
    Genre1, Genre2, Genre3, Collection_in_million, Votes, Certificate FROM imdb_2 WHERE ratings != '' ORDER BY CAST(ratings AS REAL) DESC
    LIMIT 1"""

# Execute the queries and fetch the results
cur.execute(query_imdb1)
result_imdb1 = cur.fetchone()

cur.execute(query_imdb2)
result_imdb2 = cur.fetchone()

# Compare the highest rating from both tables
highest_rating_row = None
if result_imdb1 and result_imdb2:
    if float(result_imdb1[4]) > float(result_imdb2[4]):
        highest_rating_row = result_imdb1
    else:
        highest_rating_row = result_imdb2
elif result_imdb1:
    highest_rating_row = result_imdb1
elif result_imdb2:
    highest_rating_row = result_imdb2

# Check if there is any movie with the highest rating
if highest_rating_row:
    # Print the details of the movie with the highest rating
    print("Movie Name:", highest_rating_row[0])
    print("Duration:", highest_rating_row[1])
    print("Director:", highest_rating_row[2])
    print("Year:", highest_rating_row[3])
    print("Ratings:", highest_rating_row[4])
else:
    print("No movie with the highest rating found.")
```

```python
#Details of movies with lowest rating
# Query to select movie details from imdb_1 table with valid ratings
query_imdb1 = """SELECT movie_names, duration, director_1, year, ratings, metascore, '' as Genre1, '' as Genre2, '' as Genre3,
    '' as Collection_in_million, '' as Votes, '' as Certificate FROM imdb_1 WHERE ratings != '' ORDER BY CAST(ratings AS REAL) ASC
    LIMIT 2 OFFSET 1"""

# Query to select movie details from imdb_2 table with valid ratings
query_imdb2 = """SELECT Movies as movie_names, '' as duration, '' as director_1, '' as year,
    '' as ratings, '' as metascore, Genre1, Genre2, Genre3, CAST(Collection_in_million AS REAL),
    CAST(NULLIF(Votes, '') AS INTEGER) AS Votes,Certificate FROM imdb_2
    WHERE ratings != '' ORDER BY CAST(Collection_in_million AS REAL) ASC LIMIT 2"""

# Execute the queries and fetch the results
cur.execute(query_imdb1)
result_imdb1 = cur.fetchall()

cur.execute(query_imdb2)
result_imdb2 = cur.fetchall()

# Compare the highest rating from both tables
lowest_rating_row = None
if result_imdb1 and result_imdb2:
    if float(result_imdb1[0][4]) > float(result_imdb2[0][4]):
        lowest_rating_row = result_imdb1[0]
    else:
        lowest_rating_row = result_imdb2[0]
elif result_imdb1:
    lowest_rating_row = result_imdb1[0]
elif result_imdb2:
    lowest_rating_row = result_imdb2[0]

# Check if there is any movie with the highest rating
if lowest_rating_row:
    # Print the details of the movie with the highest rating
    print("Movie Name:", lowest_rating_row[0])
    print("Duration:", lowest_rating_row[1])
    print("Director:", lowest_rating_row[2])
    print("Year:", lowest_rating_row[3])
    print("Ratings:", lowest_rating_row[4])
else:
    print("No movie with the highest rating found.")
```
