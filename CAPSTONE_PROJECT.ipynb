# Scraping of data

```python
#importing necessary libraries
import selenium
import pandas as pd
import bs4
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, TimeoutException
from selenium.common.exceptions import SessionNotCreatedException,ElementNotInteractableException
import re
import requests
from selenium.webdriver.common.by import By
import numpy as np
import warnings
warnings.filterwarnings("ignore")
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")   #Opening new page
```

```python
#Maximizing the window
driver.maximize_window()
```

```python
url="https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1"
```

```python
#opening the url
driver.get(url)
```

```python
scraped_sno=[]
start=1
end=36
for page in range(start,end):
    sno=driver.find_elements(By.XPATH,"//span[@class='lister-item-index unbold text-primary']")
    for i in sno:
        scraped_sno.append(i.text.replace(".",""))
    next_button=driver.find_element(By.XPATH,"//a[@class='lister-page-next next-page']")
    next_button.click()
    time.sleep(5)
sno2=driver.find_elements(By.XPATH,"//span[@class='lister-item-index unbold text-primary']")
for i in sno2:
    scraped_sno.append(i.text.replace(".",""))
print(scraped_sno)
```

```python
movie_names=[]
start=1
end=36
for i in range(start,end):
    names=driver.find_elements(By.XPATH,"//h3[@class='lister-item-header']//a")
    for i in names:
        movie_names.append(i.text)
    previous_button=driver.find_element(By.XPATH,"//a[@class='lister-page-prev prev-page']")
    previous_button.click()
    time.sleep(5)
names2=driver.find_elements(By.XPATH,"//h3[@class='lister-item-header']//a")
for i in names2:
    movie_names.append(i.text)
print(movie_names)
```

```python
director1=[]
start=1
end=36
for i in range(start,end):
    tags=driver.find_elements(By.XPATH,"//p[3][@class]//a[1]")
    for i in tags:
        director1.append(i.text)
    next_button=driver.find_element(By.XPATH,"//a[@class='lister-page-next next-page']")
    next_button.click()
tags2=driver.find_elements(By.XPATH,"//p[3][@class]//a[1]")
for i in tags2:
    director1.append(i.text)
print(director1)
```

```python
runtime=[]
start=1
end=36
for i in range(start,end):
    tags=driver.find_elements(By.XPATH,"//span[@class='runtime']")
    for i in tags:
        runtime.append(i.text)
    previous_button=driver.find_element(By.XPATH,"//a[@class='lister-page-prev prev-page']")
    previous_button.click()
tags2=driver.find_elements(By.XPATH,"//span[@class='runtime']")
for i in tags2:
    runtime.append(i.text)
print(runtime)
```

```python
year=[]
start=1
end=36
for i in range(start,end):
    tags=driver.find_elements(By.XPATH,"//span[@class='lister-item-year text-muted unbold']")
    for i in tags:
        year.append(i.text.replace("()",""))
    next_button=driver.find_element(By.XPATH,"//a[@class='lister-page-next next-page']")
    next_button.click()
tag2=driver.find_elements(By.XPATH,"//span[@class='lister-item-year text-muted unbold']")
for i in tag2:
    year.append(i.text)
print(year)
```

```python
ratings=[]
start=1
end=36
for i in range(start,end):
    tag=driver.find_elements(By.XPATH,"//div[1][strong]")
    for i in tag:
        ratings.append(i.text)
    previous_button=driver.find_element(By.XPATH,"//a[@class='lister-page-prev prev-page']")
    previous_button.click()
tag2=driver.find_elements(By.XPATH,"//div[1][strong]")
for i in tag2:
    ratings.append(i.text)
print(ratings)
```

```python
metascore=[]
start=1
end=36
for i in range(start, end):
    tags=driver.find_elements(By.XPATH,"//div[3][@class='inline-block ratings-metascore']")
    for i in tags:
        metascore.append((i.text.strip("Metascore").replace(" ",""))or None)
        next_button=driver.find_element(By.XPATH,"//a[@class='lister-page-next next-page']")
    next_button.click()
tag2=driver.find_elements(By.XPATH,"//div[3][@class='inline-block ratings-metascore']/div[3]/span")
for i in tag2:
    metascore.append((i.text.strip())or None)
print(metascore)
```

```python
metascore.extend(["-"]*(1770-len(metascore)))
```

```python
print(len(scraped_sno),len(movie_names),len(runtime),len(director1),len(year),len(ratings),len(metascore))
```

```python
df1=pd.DataFrame()
df1['S.No.']=scraped_sno
df1['Movie_Names']=movie_names
df1['Duration']=runtime
df1['Director_1']=director1
df1['Year']=year
df1['ratings']=ratings
df1['Metascore']=metascore
df1.sample(10)
```

```python
df1.to_csv("imdb1.csv")
```
