```python
base_url = "https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1"
```

```python
# Total number of movies
total_movies = 1770

# Movies per page
movies_per_page = 50

# Specify the start and end page numbers
start_page = 1
end_page = 36
data_genre=[]
for page in range(start_page, end_page + 1):
    # Construct the URL for the current page
    url = f"{base_url}&start={(page - 1) * movies_per_page}"
    driver.get(url)

    # Extract the HTML content after the page has fully loaded
    html = driver.page_source

    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')

    # Find the genre elements using BeautifulSoup
    genre_elements = soup.find_all('span', class_='genre')

    # Extract the genres for each element
    for genre_element in genre_elements:
        genre_text = genre_element.get_text(strip=True)
        genres = genre_text.split(',')
        genres = [genre.strip() for genre in genres]
        
        # Extract genre1, genre2, and genre3
        genre1 = genres[0].strip() if len(genres) > 0 else None
        genre2 = genres[1].strip() if len(genres) > 1 else None
        genre3 = genres[2].strip() if len(genres) > 2 else None
        
        data_genre.append({"Genre1":genre1,"Genre2":genre2,"Genre3":genre3})   #appending the data into the empty list
```

```python
start=1
end=36
star1 = []
for i in range(start,end):
    movie_items = driver.find_elements(By.XPATH, '//div[@class="lister-item-content"]')
    for item in movie_items:
        elements = item.find_elements(By.XPATH, './/p[contains(text(), "Director:")]/a')
        if len(elements) >= 2:
            star = elements[1].text.strip()
            star1.append(star)
        else:
            star1.append("-")
    next_button=driver.find_element(By.XPATH,"//a[@class='lister-page-next next-page']")
    next_button.click()
movie_items2 = driver.find_elements(By.XPATH, '//div[@class="lister-item-content"]')
for item in movie_items2:
    elements = item.find_elements(By.XPATH, './/p[contains(text(), "Director:")]/a')
    if len(elements) >= 2:
        star = elements[1].text.strip()
        star1.append(star)
    else:
        star1.append("-")
print(star1)
```

```python
# Pad list with None values to match other columns
star1.extend([None] * (1771 - len(star1)))
```

```python
# Total number of movies
total_movies = 1765

# Movies per page
movies_per_page = 50

# Specify the start and end page numbers
start_page = 1
end_page = 36
data_stars=[]
for page in range(start_page, end_page + 1):
    # Construct the URL for the current page
    url = f"{base_url}&start={(page - 1) * movies_per_page}"
    driver.get(url)

    # Extract the HTML content after the page has fully loaded
    html = driver.page_source

    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')
    star_tag=soup.find_all("div",class_='lister-item-content')
    for i in star_tag:
        star_list=i.find("p",class_='').find_all("a")
        stars=[i.get_text(strip=True)for i in star_list]
        data_stars.append({"Star"+str(i+1):stars[i]if i<len(stars) else None  for i in range(1,4)})   #separating the stars as 1, 2,3 and so on..
```

```python
df1=pd.DataFrame(data_stars)
df1
```

```python
votes=[]
start=1
end=36
for i in range(start,end):
    votes_tag=driver.find_elements(By.XPATH,"//p[@class='sort-num_votes-visible']//span[2]")
    for i in votes_tag:
        votes.append(i.text.strip().replace(",",""))
    next_button=driver.find_element(By.XPATH,"//a[@class='lister-page-prev prev-page']")
    next_button.click()
tag2=driver.find_elements(By.XPATH,"//p[@class='sort-num_votes-visible']//span[2]")
for i in tag2:
    votes.append(i.text.strip().replace(",",""))    #appending the scraped data of last page
print(votes)
```

```python
certificate=[]    #creating empty list
start=1
end=36
for i in range(start,end):
    tag=driver.find_elements(By.XPATH,"//span[@class='certificate']")
    for i in tag:
        certificate.append(i.text.strip())
    next_button=driver.find_element(By.XPATH,"//a[@class='lister-page-next next-page']")
    next_button.click()
tag2=driver.find_elements(By.XPATH,"//span[@class='certificate']")
for i in tag2:
    certificate.append(i.text.strip())     #appending the scraped data from last page
print(certificate)
```

```python
certificate.extend(["-"]*(1771-len(certificate)))    #padding to match the total number of rows with other columns
```

```python
collection = []
start = 1
end = 36

for i in range(start, end):
    tag = driver.find_elements(By.XPATH, "//p[@class='sort-num_votes-visible']//span[5]")
    for i in tag:
        symbols = ['$', 'M']
        value = i.text.strip()
        for i in symbols:
            value = value.replace(i, "")
        collection.append(value or None)  # Append None if value is missing or empty
        
    previous_button = driver.find_element(By.XPATH, "//a[@class='lister-page-prev prev-page']")
    previous_button.click()

tag2 = driver.find_elements(By.XPATH, "//p[@class='sort-num_votes-visible']//span[5]")
for i in tag2:
    symbols = ['$', 'M']
    value = i.text.strip()
    for i in symbols:
        value = value.replace(i, "")
collection.append(value or None)  # Append None if value is missing or empty

print(collection)
```

```python
# Pad collection list with None values to match other columns
collection.extend([None] * (1771 - len(collection)))
```

```python
movies=[]
start=1   #giving the starting page
end=36   #last page
for i in range(start,end):
    tag=driver.find_elements(By.XPATH,"//h3[@class='lister-item-header']//a")
    for i in tag:
        movies.append(i.text.strip())
    next_button=driver.find_element(By.XPATH,"//a[@class='lister-page-next next-page']")  #Automting opening of next page
    next_button.click()
tag2=driver.find_elements(By.XPATH,"//h3[@class='lister-item-header']//a")
for i in tag2:
    movies.append(i.text.strip())
print(movies)
```

```python
movies.extend([None] * (1771 - len(movies)))
```

```python
#creating a dataset from the scraped info
df=pd.DataFrame(data_genre)
df['Movies']=movies
df['Collection(in million)']=collection
df['Votes']=votes
df['Certificate']=certificate
df['Star1']=star1
df.sample(10)
```
