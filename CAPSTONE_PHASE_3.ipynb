# Outliers Detection

```python
#We will only remove outliers from numerical columns
for col in num_col:
    plt.figure(figsize=(4,4))
    sns.boxplot(data=df, y=col, color='purple')
    plt.ylabel(col, fontsize=12)
    plt.title(f'Boxplot for {col}', fontsize=15)
    plt.show()
```

We can see there are lots of outliers present in the dataset. But as the 'Votes' and 'Collection(in million)' are target varibles, we will not remove any outliers from them, only lets remove outliers from other columns.

```python
#importing the library 
from scipy.stats import zscore
```

```python
outliers=df[['ratings','Duration','Metascore']]   #creating a list for columns having outliers
z=np.abs(zscore(outliers)) #Applying the zscore
z  #Displaying the output
```

```python
#Checking the rows where zscore is greater than threshold value
threshold=3
print(np.where(z>3))
```

```python
#creating a new dataframe without the outliers
df_new=df[(z<3).all(axis=1)]
df_new
```

```python
#Checking data loss after removing the outliers
print("Data loss percentage=",(df.shape[0]-df_new.shape[0])/df.shape[0]*100)
```

As the data loss is only less, we can definitely proceed with the new dataset.

```python
df=df_new
print("The shape of new dataset is:",df.shape)
```

# Target_Encoding

```python
#Since there are certain columns in the features like 'Director_1' which can't be encoded using label encoder,
#and also can't be dropped, we will use the target encoding method to retain the useful information from these columns.

#First we will encode DIRECTOR_1

#Lets first calculate the mean of the target variables for each unique 'Director_1'
mean_collection=df.groupby('Director_1')['Collection(in million)'].mean()  #First target variable's mean
mean_votes=df.groupby('Director_1')['Votes'].mean() #2nd target variables' mean

#creating new encoded Director_1 columns
x['Director_Collection']=x['Director_1'].map(mean_collection)
x['Director_Votes']=x['Director_1'].map(mean_votes)

#Dropping the original column
x.drop('Director_1',axis=1,inplace=True)
x.head()
```

```python
#Encoding the Star1 column
mean_collection=df.groupby('Star1')['Collection(in million)'].mean()
mean_votes=df.groupby('Star1')['Votes'].mean()

x['Star1_collection']=x['Star1'].map(mean_collection)
x['Star1_votes']=x['Star1'].map(mean_votes)

#Dropping the original column
x.drop('Star1',axis=1,inplace=True)
x.head()  #checking the output
```

```python
#Encoding Star2 column
mean_collection=df.groupby('Star2')['Collection(in million)'].mean()
mean_votes=df.groupby('Star2')['Votes'].mean()

#Mapping the encoded values to new columns
x['Star2_collection']=x['Star2'].map(mean_collection)
x['Star2_votes']=x['Star2'].map(mean_votes)

#Dropping the original column
x.drop('Star2',axis=1,inplace=True)
x.head()   #Checking the output
```

```python
#Encoding Star3 column
mean_collection=df.groupby('Star3')['Collection(in million)'].mean()
mean_votes=df.groupby('Star3')['Votes'].mean()

x['Star3_collection']=x['Star3'].map(mean_collection)
x['Star3_votes']=x['Star3'].map(mean_collection)

#Dropping the actual column
x.drop('Star3',inplace=True,axis=1)
#Checking the output
x.head()
```

```python
#Encoding the Star4
mean_collection=df.groupby('Star4')['Collection(in million)'].mean()
mean_votes=df.groupby('Star4')['Votes'].mean()

#Mapping the encoded values
x['Star4_collection']=x['Star4'].map(mean_collection)
x['Star4_votes']=x['Star4'].map(mean_votes)

#Dropping the actual column
x.drop('Star4',inplace=True,axis=1)

#checking the output
x.head()
```

# Multi_Collinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor   #importing the VIF library
```

```python
#We will only select numerical columns 
num_col=x.select_dtypes(include=['int64','float64']).columns  #creating a list of numerical columns
x_num=x[num_col] #creating a data frame having the numerical columns
vif_data=pd.DataFrame()
vif_data['VIF_values']=[variance_inflation_factor(x_num.values,i)for i in range(x_num.shape[1])]
vif_data['Features']=x_num.columns
vif_data
```
