```python
#importing the libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

# TABLE_1

```python
#importing the csv file
df1=pd.read_csv("IMDB_1.csv")
df1.head()
```

```python
#Display all the details of movies created by directors Christopher Nolan and Matt Reeves.
directors=['Christopher Nolan','Matt Reeves']  #creating a list of selected directors
movies=df1.loc[df1['Director_1'].isin(directors)]   #locating the details
movies     #displaying the output
```

```python
# Extract numeric value from the "Duration" column using regex and convert to float
df1['Duration'] = df1['Duration'].str.extract('(\d+)').astype(float)

movies_within_duration_range = df1.loc[(df1['Duration'] >= 140) & (df1['Duration'] <= 190).values]

movies_within_duration_range
```

```python
# Display all details of movies with ratings above 7 in ascending order.
movies=df1.loc[df1['ratings']>=7]
sorted_movies=movies.sort_values(by='ratings',ascending=True)
sorted_movies
```

```python
#Display all movie names in descending order.
sorted_movies=df1.sort_values(by='Movie_Names',ascending=False)
movies=sorted_movies['Movie_Names']
print(movies)
```

```python
#Display movie name starts with ‘P’ and their rating is greater than 7.
movies_filtered = df1[(df1['Movie_Names'].str.startswith('P')) & (df1['ratings'] > 7)]
(movies_filtered)
```

# TABLE_2

```python
#importing the csv file
df2=pd.read_csv("IMDB_2.csv")
df2.head()
```

```python
#Display all movie names with star Arnold Schwarzenegger in ascending order.
#creating a condition for finding the movies in all the 4 star columns
arnold=df2[['Star1','Star2','Star3','Star4']].apply (lambda x: x.str.contains('Arnold Schwarzenegger',case=False)).any(axis=1)
movies=df2[arnold]
sorted_movies=movies.sort_values(by='Movies',ascending=True)  #sorting the movies in ascending order
sorted_movies[['Movies','Star1','Star2','Star3','Star4']]  #displaying movie names of Arnold Schwarzenegger
```

```python
#Display all details of the movie with the highest number of votes
highest_votes=df2['Votes'].idxmax() #Displaying the index having maximum votes
movie=df2.loc[highest_votes]   #fetching the details of movie having highest index
print(movie)   #Displaying the output
```

```python
#Display movie names with gross collections in descending order.
sorted_movies=df2.sort_values(by='Collection(in million)',ascending=False)  #creating the condition
sorted_movies.head(5)  #Displaying the output
```

```python
# Display the gross collection of movies with the star Arnold
arnold=df2[['Star1','Star2','Star3','Star4']].apply(lambda x: x.str.contains('Arnold Schwarzenegger',case=False)).any(axis=1)
movies=df2[arnold]   #collection movies of 'ARNOLD'
movies[['Movies','Collection(in million)']]   #displaying the output
```

```python
#Display all details of movies with comedy and action genres.
#creating instance for finding the details of movies
comedy_action= ((df2[['Genre1', 'Genre2', 'Genre3']].apply(lambda col: col.str.contains('Comedy', case=False)).sum(axis=1) > 0)
    & (df2[['Genre1', 'Genre2', 'Genre3']].apply(lambda col: col.str.contains('Action', case=False)).sum(axis=1) > 0))

#fetching the details of movies having both action and comedy genre
movies=df2[comedy_action]

#Displaying the output
movies   
```
